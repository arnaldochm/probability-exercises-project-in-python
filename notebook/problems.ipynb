### Exercise 1 

Two unbiased dice are thrown once and the total score is observed. Use a simulation to find the estimated probability that the total score is even or greater than 7

We could use a mathematical equation to solve this or we can run the experiment many times, and then count the number of times where the score is even or greater than 7

-Run the experiment 1000 times (roll 2 dice 1000 times, and sum the result)

-Keep track of the number of times that the sum was either greater than 7 or an even number

-Divide the number from step 2 by the number of iterations (1000)


```python
#import libraries
import numpy as np
import pandas as pd
import random

#Function to roll the dice.
def roll_dice(experiment_retries):

    even_or_seven = 0
    for _ in range(0, experiment_retries):

        #Dice 1 possible events: {1,2,3,4,5,6}
        dice1 = random.randint(1,6)

        #Dice 2 possible events: {1,2,3,4,5,6}
        dice2 = random.randint(1,6)

        result = dice1 + dice2

        if (result == 7 or result%2 == 0):
            even_or_seven += 1

    return even_or_seven/experiment_retries

print(f'The total probability that the score is even or 7 is: {roll_dice(1000)}')
```

### Exercise 2

A box contains 10 white balls, 20 reds and 30 greens. Draw 5 balls with replacement. What is the probability that:

a. 3 white or 2 red

b. All 5 are the same color


We will use the same system here. We will pick our balls during each round and count the number of times that:

a. 3 white or 2 red 

b. All 5 are the same color.

```python
#import libraries

d = {}

for i in range(60):
    if i < 10:
        d[i] = 'White'
    elif i > 9 and i < 30:
        d[i] = 'Red'
    else:
        d[i] = 'Green'  
          
#Initialize important variables
n_simulations = 10000
w_and_r_count = 0
same_color = 0

for i in range(n_simulations):
    
    array_result = []

    #make a list of the colors that we choose
    random_result = np.random.choice(np.arange(0,60,1), 5)
    
    #Convert it to a numpy array
    for random_val in random_result:
        array_result.append(d[random_val])

    array_result = np.array(array_result)

    #find the number of each taht we picked
    red = sum(array_result == 'Red')
    white = sum( array_result == 'White')
    green = sum(array_result =='Green')
    
    #print(f'red={red}, white = {white} and green = {green}')

    #Keep track if the combination met the criteria
    if (white == 3 and red == 2):
        w_and_r_count += 1
    
    # If two elements from red,white,green are zero, it means all 5 results where of the same color
    if (red == 5 or white == 5 or green == 5): 
        same_color += 1
    
#print 'the probability of 3 white and 2 red is' in percentage
print(f'the probability of 3 white and 2 red is {(w_and_r_count/n_simulations)*100}')

#print 'the probability of all the same color is' in percentage   
print(f'the probability of all the same color is {(same_color/n_simulations)*100}')
```
